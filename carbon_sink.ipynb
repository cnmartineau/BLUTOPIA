```python
# import libraries
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
fig = go.Figure()
config = {'displayModeBar': True}

# plot algae quantities
values = [75, 82, 101, 109, 151, 83, 174, 210, 93]
labels = [
    "Forêts de kombu<br>(<i>Ecklonia</i>)",
    "Forêts de kombu<br>(<i>Nereocystis</i>)",
    "Forêts de kombu<br>(<i>Macrocystis</i>)",
    "Forêts de kombu<br>(<i>Laminaria/Saccharina</i>)",
    "Forêts de kombu<br>(<i>Lessonia</i>)",
    "Forêts d'herbes marines",
    "Mangroves",
    "Marais salants",
    "Forêts terrestres",
]
fig.add_trace(
    go.Bar(
        x=values,
        y=labels,
        text=values,
        textposition="inside",
        textfont=dict(color="#93C9F9"),
        marker=dict(
            cornerradius=30,
            color="#113972"
        ),
        orientation="h",
        hoverinfo="skip",
        showlegend=False
    ),
)

fig.update_xaxes(
    showgrid=True,
    color="#93C9F9",
    tickvals=np.arange(0, 400, 100),
    range=[0, 450],
    gridcolor="#93C9F9"
)

fig.update_yaxes(
    showline=True,
    color="#113972",
    linecolor="#93C9F9",
    tickangle=0,
    ticklabelstandoff=10,
    autorange="reversed"
)

fig.update_layout(
    plot_bgcolor="white",
    hovermode="closest",
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.3,
        xanchor="center",
        x=0.5
    ),
    margin=dict(
        t=20,
        b=20,
        l=20,
        r=20
    ),
    modebar=dict(
        orientation="h",
    ),
    height=500,
    width=600
)

fig.show(config=config)
```

```python
fig.write_html("carbon_sinks.html", include_plotlyjs="cdn")
```
