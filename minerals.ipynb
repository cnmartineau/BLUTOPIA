```python
# import libraries
import pandas as pd
import plotly.graph_objects as go
```

```python
# load data
data = pd.read_csv("minerals.csv", sep=";")
```

```python
# preview data
data.head()
```

```python
# drop unused data
data = data.drop([15], axis=0).reset_index(drop=True)
data = data.drop(["Chlorures"], axis=1)
```

```python
# format vernacular name
mask_nan = data["Nom commun"].isnull()
data.loc[mask_nan, "Nom commun"] = ""
```

```python
# format numbers from strings to float
columns = data.columns[4:]
for column in columns:
    mask_ok = data[column].notnull()
    data.loc[mask_ok, column] = [float(strnb.replace(",", ".")) for strnb in data.loc[mask_ok, column]]
```

```python
fig = go.Figure()
config = {'displayModeBar': True}

customscale = [[ 0.0, "#CCE9FF"], [1.0, "#007DED"]]

fig.add_trace(
    go.Heatmap(
        z=data[columns],
        x="<b>" + columns + "</b>",
        y="""<span style="color:""" + data["Couleur"] + """"><b>""" + \
            data["Nom commun"] + "</b><br><i>" + data["Nom scientifique"] + """</i></span>""",
        text=data[columns].astype(str),
        texttemplate="%{text}",
        textfont=dict(size=10),
        colorscale=customscale,
        showscale=False,
        hoverinfo="skip"
    ),
)

fig.update_xaxes(
    showgrid=False,
    side="top",
    tickangle=0,
    ticklabelstandoff=10,
    color="#007DED"
),

fig.update_yaxes(
    showgrid=False,
    tickangle=0,
    ticklabelstandoff=10,
    autorange="reversed"
)

fig.update_layout(
    plot_bgcolor="white",
    margin=dict(
        t=20,
        b=20,
        l=20,
        r=20
    ),
    modebar=dict(
        orientation="h",
    ),
    height=800,
    width=1200
)

fig.show(config=config)
```

```python
fig.write_html("minerals.html", include_plotlyjs="cdn")
```
