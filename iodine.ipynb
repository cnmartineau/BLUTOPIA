```python
# import libraries
import pandas as pd
import plotly.graph_objects as go
```

```python
# load data
data = pd.read_csv("minerals.csv", sep=";")
```

```python
# preview data
data.head()
```

```python
# drop unused data
data = data.drop([15], axis=0).reset_index(drop=True)
data = data.drop(["Chlorures"], axis=1)
```

```python
# format vernacular name
mask_nan = data["Nom commun"].isnull()
data.loc[mask_nan, "Nom commun"] = ""
```

```python
# format numbers from strings to float
columns = data.columns[4:]
for column in columns:
    mask_ok = data[column].notnull()
    data.loc[mask_ok, column] = [round(float(strnb.replace(",", ".")), 0) for strnb in data.loc[mask_ok, column]]
```

```python
fig = go.Figure()
config = {'displayModeBar': True}

fig.add_trace(
    go.Bar(
        x=data["Iode"],
        y="""<span style="color:""" + data["Couleur"] + """"><b>""" + \
            data["Nom commun"] + "</b><br><i>" + data["Nom scientifique"] + """</i></span>""",
        text=data["Iode"],
        textposition="outside",
        textfont=dict(color="#93C9F9"),
        marker=dict(
            cornerradius=30,
            color="#93C9F9"
        ),
        orientation="h",
        hoverinfo="skip"
    ),
)

fig.update_xaxes(
    title=dict(
        text="Concentration en iode (en mg par 100g d'algue déshydratée)",
        font=dict(color="#93C9F9")
    ),
    showline=False,
    showgrid=True,
    showticklabels=False,
    range=[0,510],
    gridcolor="#93C9F9"
),

fig.update_yaxes(
    showline=True,
    linecolor="#93C9F9",
    tickangle=0,
    ticklabelstandoff=10,
    autorange="reversed"
)

fig.update_layout(
    plot_bgcolor="white",
    barcornerradius=15,
    margin=dict(
        t=20,
        b=50,
        l=20,
        r=20
    ),
    modebar=dict(
        orientation="h",
    ),
    height=800,
    width=800
)

fig.show(config=config)
```

```python
fig.write_html("iodine.html", include_plotlyjs="cdn")
```
